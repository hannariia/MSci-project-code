```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# pick df
df = pd.read_csv("/home/hra35/Documents/events/nll_script_outputs/D7/picks_final.txt", sep=r'\s+', header=0)
```

```python
# P-waves
p_df = df[df["Phase"] == "P"]

# average picking error
perr_avg = np.average(p_df["TT_err"])
print("Average picking error for P-waves is ", perr_avg, "s")

# error bins
p_bins = list(p_df['TT_err'].unique())
p_bins.sort()
p_err_counts = p_df['TT_err'].value_counts()

# bar plot
plt.figure(figsize=(10, 6))
p_err_counts[p_bins].plot(kind='bar', edgecolor='black', color="red", width=0.8)

plt.title('P-arrival picking errors', fontsize=22)
plt.xlabel('Reported picking error (s)', fontsize=22)
plt.ylabel('Number of picks', fontsize=22)
plt.xticks(fontsize=22, rotation=0)
plt.yticks(fontsize=22)

plt.savefig("/raid4/hra35/gmt_figures/report/P_errors.png", facecolor='white',  bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# S-waves

s_df = df[df["Phase"] == "S"]

# average picking error
serr_avg = np.average(s_df["TT_err"])
print("Average picking error for S-waves is ", serr_avg, "s")

# error bins
s_bins = list(s_df['TT_err'].unique())
s_bins.sort()
s_err_counts = s_df['TT_err'].value_counts()

# bar plot
plt.figure(figsize=(10, 6))
s_err_counts[s_bins].plot(kind='bar', edgecolor='black', color="red", width=0.8)

plt.title('S-wave arrival picking errors', fontsize=22)
plt.xlabel('Reported picking error (s)', fontsize=22)
plt.ylabel('Number of picks', fontsize=22)
plt.xticks(fontsize=22, rotation=0)
plt.yticks(fontsize=22)

plt.savefig("/raid4/hra35/gmt_figures/report/S_errors.png", facecolor='white', bbox_inches='tight')

plt.show()
```
